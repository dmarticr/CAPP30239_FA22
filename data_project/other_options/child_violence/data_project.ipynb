```python
import urllib
import pandas as pd
import geopandas as gpd
import numpy as np

# Sources:
# - https://micrositios.centrodememoriahistorica.gov.co/observatorio/portal-de-datos/base-de-datos/
# - https://www.kaggle.com/code/alfredomaussa/colombia-municipios

outfilename = "dta.xlsx"
dls = "https://micrositios.centrodememoriahistorica.gov.co/observatorio/download/victimas-ru/?wpdmdl=1272&refresh=635cc5f7319b51667024375"
urllib.request.urlretrieve(dls, outfilename)
dta = pd.read_excel(outfilename)
#dta_geo = gpd.read_file("Municipios.json")
```

```python
col_names = {"ID Caso": "ID",
            "Código DANE de Municipio": "DPTOMPIO",
            "Municipio": "MPIO_CNMBR",
            "Departamento": "DEPARTMENT",
            "Año": "YEAR",
            "Mes": "MONTH",
            "Día": "DAY",
            "ID Persona": "ID",
            "Sexo": "GENDER",
            "Etnia": "ETHNICITY",
            "Ocupación":"OCCUPATION",
            "Caletera":"ROBBER",
            "Campanera":"ALARM",
            "Cocinera":"COOK",
            "Comandante":"COMMANDER",
            "Combatiente":"FIGHTER",
            "Contabilidad":"ACCOUNTING",
            "Entrenamiento":"TRAINING",
            "Escolta":"SHOOTING GUARD",
            "Extorsiones":"EXTORTION",
            "Fabricación de Otras Armas No Convencionales":"WEAPON MANUFACTURING",
            "Fabricación y/o Siembra de Minas Antipersonal":"MANUFACTURE-PLANTING OF MINES",
            "Guardia":"GUARD",
            "Informante":"INFORMANT",
            "Patrullera":"PATROLMAN",
            "Radio Operaciones":"RADIO OPERATIONS",
            "Raspachín":"COCA PICKER",
            "Servicios de Salud":"HEALTH SERVICES",
            "Sicariato":"CONTRACT KILLER",
            "Trabajo Organizativo":"ORGANIZATIONAL WORK",
            "Tráfico de Estupefacientes":"DRUG TRAFFICKING",
            "Transporte de Armas":"WEAPONS TRANSPORT",
            "Sin Información de Oficios Realizados":"NO INFO",
            "Otro Oficio Realizado":"OTHER",
            "Edad":"AGE",
            "Latitud": "LAT",
            "Longitud": "LON"}

sex_dic = {"HOMBRE":"MALE",
            "MUJER":"FEMALE",
            "SIN INFORMACION":"NO INFO"}

eth_dic = {"INDÍGENA":"INDIGENOUS",
            "AFROCOLOMBIANO":"AFRO-COLOMBIAN",
            "ROM":"GYPSY"}

occ_dic = {"SIN INFORMACIÓN":"NO INFORMATION",
            "GUERRILLERO":"GUERRILLA",
            "ESTUDIANTE":"STUDENT",
            "OBRERO":"WORKER",
            "CAMPESINO":"FARMER",
            "PARAMILITAR":"PARAMILITARY",
            "RASPACHÍN":"COCA PICKER",
            "ECONOMÍA INFORMAL":"INFORMAL ECONOMY",
            "TRABAJADOR DE FINCA":"FARM WORKER",
            "NO APLICA":"DOES NOT APPLY",
            "EMPLEADO":"EMPLOYEE",
            "OTRA ¿CUÁL?":"OTHER",
            "MINERO":"MINER",
            "COMERCIANTE":"TRADE",
            "TRABAJO SIN ESPECIFICAR":"UNSPECIFIED WORK",
            "MIEMBRO DE GRUPO POSDESMOVILIZACIÓN":"POST-DEMOBILIZATION GROUP MEMBER",
            "AMA DE CASA":"HOUSEWIFE"}

age_dic = {"INFANCIA (6-11 AÑOS)":"INFANCIA (6-11)",
            "ADOLESCENCIA (12-17 AÑOS)":"ADOLESCENCIA (12-17)",
            "PRIMERA INFANCIA (0-5 AÑOS)":"PRIMERA INFANCIA (0-5)"}

dta.rename(columns=col_names, inplace=True)
dta = dta.replace({"GENDER": sex_dic})
dta = dta.replace({"ETHNICITY": eth_dic})
dta = dta.replace({"AGE": age_dic})
dta = dta.replace({"OCCUPATION": occ_dic})
#dta = dta.replace(r'^\s*$', np.nan, regex=True)
dta.head()
```
