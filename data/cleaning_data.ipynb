```python
import pandas as pd
import numpy as np

#dictionary of firms
dct = pd.read_csv('dict.csv', index_col = 'Platform')
dct = dct.to_dict()

df = pd.read_csv('vgsales2.csv')
df['Year'] = df['Year'].replace('N/A', np.nan)
num_cols = ['NA_Sales', 'PAL_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Total_Shipped', 'Sales']
for col in num_cols:
    df[col] = df[col].fillna(0)
df['Sales'] = df['Global_Sales'] + df['Total_Shipped']
df['Year'] = df['Year'].astype('Int64')
df['Company'] = df['Platform'].map(dct["Mother"]).fillna('Other')
df['Year_console'] = df['Platform'].map(dct["Year"]).astype('Int64')
df['Name_platform'] = df['Platform'].map(dct["Name"]).fillna(df['Platform'])
df
```

# 1) Total sales from Nintendo and others by year

```python
# for this case, I need the year so I remove if not in dataset
g11 = df[df['Year'].notna()]
g11 = g11.groupby(['Year', "Company"])['Sales'].sum().reset_index()
#g11.to_csv('g11.csv', index = False)

g11 = g11.set_index(['Year', 'Company']).unstack(['Company'], fill_value = 0)
g11.columns = g11.columns.get_level_values(1)
g11.to_csv('g11.csv')
g11
```

```python
base = df[["Company","Name_platform","Name","Sales"]]
names = base.groupby(["Name"])['Sales'].sum().to_dict()
plat = base[["Name_platform","Name"]].drop_duplicates().to_dict('records')
platform = {}
for dct in plat:
    plt = dct['Name_platform']
    platform[plt] = platform.get(plt, []) + [dct['Name']]
platform
company = {}
comp = base[["Company", "Name_platform"]].drop_duplicates().to_dict('records')
for dct in comp:
    plt = dct['Company']
    company[plt] = company.get(plt, []) + [dct['Name_platform']]
company
final = {}
for comp_key, list_plat in company.items():
    final[comp_key] = {}
    for plat in list_plat:
        temp = {}
        temp[plat] = {}
        list_names = platform[plat]
        for name in list_names:
            temp[plat][name] = names[name]
        final[comp_key][plat] = temp[plat]

final
```

```python
def retro(dic):
    fn = {}
    for key, value in dic.items():
        if type(value) is not dict:
            return {'name': key, 'value': value}
        fn['name'] = key, 
        fn['children'] = [retro(value)]
    return fn
            
retro(final)
```

```python
dic_final = []

for key, children in final.items():
    dic_la = {}
    dic_la["name"] = key
    dic_la['children'] = []
    for key2, val2 in children.items():
        dic_intern = {}
        dic_intern['name'] = key2
        dic_intern['children'] = [{'name': key3, 'value': value3} for key3, value3 in val2.items()]
        dic_la['children'].append(dic_intern)
    dic_final.append(dic_la)
        
videogames = {'name' : 'Videogames'}
videogames['children'] = dic_final
```

```python
import json
with open("sample.json", "w") as outfile:
    json.dump(videogames, outfile, indent = 4)
```

## Reduce number of platforms and pulishers to most popular ones

```python
pop_Platform = ['2600','DS','GB','N64','NES','PS','PS2','PS3','PS4','SNES','Wii','X360','XOne']
pop_Publisher = ["Nintendo","Electronic Arts","Activision","Sony Computer Entertainment","Ubisoft",
                  "Take-Two Interactive","THQ","Konami Digital Entertainment","Sega","Namco Bandai Games"]
pop_Publisher_add = ["Microsoft Game Studios","Capcom","Atari","Warner Bros. Interactive Entertainment",
                  "Square Enix","Microsoft Game Studios","2K Games","Blizzard Entertainment","Bethesda Softworks",
                  "Atari","2K Sports","LucasArts","Sony Interactive Entertainment"]

for i, j in [("Platform", pop_Platform), ("Publisher", pop_Publisher)]:
    df[f'{i}_condensed'] = df[i]
    df[f'{i}_condensed'] = df.apply(lambda x: "Other" if x[i].strip() not in j\
                                    else x[i], axis = 1)

df
```

```python
# Historical evolution by categories
for i in ['Genre','Platform_condensed','Publisher_condensed']:
    df1 = df.groupby(['Year',i])['Global_Sales'].sum().reset_index()
    df1 = df1[df1['Year'] < 2016]
    df1['Year'] = df1['Year'].astype('int')
    df1.to_csv(f'sales_{str.lower(str(i).split("_")[0])}.csv', index = False)
```

```python
# Historical evolution by region
df2 = df.groupby(['Year'])['NA_Sales','EU_Sales','JP_Sales','Other_Sales'].sum().reset_index()
df2 = df2[df2['Year'] < 2016]
df2['Year'] = df2['Year'].astype('int')
df2 = df2.set_index("Year").stack().to_frame().reset_index().rename(columns={'level_1': 'Region', 0: 'Sales'})
df2.to_csv('sales_region.csv', index = False)
df2
```

```python
# Historical Evolution Total
df3 = df.groupby(['Year'])['Global_Sales'].sum().reset_index()
df3 = df3[df3['Year'] < 2016]
df3['Year'] = df3['Year'].astype('int')
df3.to_csv('sales_tot.csv', index = False)
df3
```

```python
# Total ranking by Genre, Platform and Publisher
for i in ['Genre','Platform_condensed','Publisher_condensed']:
    df4 = df.groupby([i])['Global_Sales'].sum().reset_index().sort_values('Global_Sales', ascending = False)
    df4.to_csv(f'totrank_{str.lower(str(i).split("_")[0])}.csv', index = False)
```
