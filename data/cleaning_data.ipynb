```python
import pandas as pd
import numpy as np
import json

#Dictionary of firms
dct = pd.read_csv('dict.csv', index_col = 'Platform')
dct = dct.to_dict()

#Data from scrapping
df = pd.read_csv('vgsales2.csv')
df['Year'] = df['Year'].replace('N/A', np.nan)
num_cols = ['NA_Sales', 'PAL_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Total_Shipped', 'Sales']
for col in num_cols:
    df[col] = df[col].fillna(0)
df['Sales'] = df['Global_Sales'] + df['Total_Shipped']
df['Year'] = df['Year'].astype('Int64')
df['Company'] = df['Platform'].map(dct["Mother"]).fillna('Other')
df['Year_console'] = df['Platform'].map(dct["Year"]).astype('Int64')
df['Name_platform'] = df['Platform'].map(dct["Name"]).fillna(df['Platform'])

pop_Publisher = ["Nintendo","Electronic Arts","Activision","Sony Computer Entertainment","Ubisoft",
                  "Take-Two Interactive","THQ","Konami Digital Entertainment","Sega","Namco Bandai Games"]
pop_add = ["Microsoft Game Studios","Capcom","Atari","Warner Bros. Interactive Entertainment",
                  "Square Enix","Microsoft Game Studios","2K Games","Blizzard Entertainment","Bethesda Softworks",
                  "Atari","2K Sports","LucasArts","Sony Interactive Entertainment"]


df['Publisher_condensed'] = df["Publisher"]
df['Publisher_condensed'] = df.apply(lambda x: "Other" if x["Publisher"].strip() not in pop_Publisher\
                                    else x["Publisher"], axis = 1)
df
```

```python
dfnf = df
df = df[df['Name_platform'] != "Series"]
df = df[df['Name_platform'] != "All"]
df
```

# 1) Total sales from Nintendo and others by year

```python
#For this case, I need the year so I remove every obs that does not have it in the dataset
g11 = df[df['Year'].notna()]
g11 = g11.groupby(['Year', "Company"])['Sales'].sum().reset_index()
g11 = g11.set_index(['Year', 'Company']).unstack(['Company'], fill_value = 0)
g11.columns = g11.columns.get_level_values(1)
g11.to_csv('g11.csv')
g11
```

```python
#Similar for publisher, not console
g11_alt = df[df['Year'].notna()]
g11_alt = g11_alt.groupby(['Year', "Publisher_condensed"])['Sales'].sum().reset_index()
g11_alt = g11_alt.set_index(['Year', 'Publisher_condensed']).unstack(['Publisher_condensed'], fill_value = 0)
g11_alt.columns = g11_alt.columns.get_level_values(1)
g11_alt.to_csv('g11_alt.csv')
g11_alt
```

```python
def create_dic(cat_lst, center):
    A, B, C, D = cat_lst
    base = df[[A, B, C, D]]

    #each level convert to list of dictionaries for correct keys use rename
    L = base.rename(columns={C:'name', D:'value'})\
            .groupby([A, B])['name','value']\
            .apply(lambda x: x.to_dict('r'))\
            .rename('children')\
            .reset_index()\
            .rename(columns={B:'name'})\
            .groupby(A)['name','children']\
            .apply(lambda x: x.to_dict('r'))\
            .rename('children')\
            .reset_index()\
            .rename(columns={A: 'name'})\
            .to_dict('r')

    #returns outer level by contructor
    return {"name": center, "children": L}

cat_lst, center = ["Company", "Name_platform", "Name", "Sales"], 'Videogames\nby\nPlatform'
videogames = create_dic(cat_lst, center)
videogames
```

```python
with open("tree.json", "w") as outfile:
    json.dump(videogames, outfile, indent = 4)
```

```python
cat_lst2, center2  = ["Publisher_condensed", "Company", "Name", "Sales"], 'Videogames\nby\nPublisher'
videogames2 = create_dic(cat_lst2, center2)
videogames2
with open("tree_alt.json", "w") as outfile:
    json.dump(videogames2, outfile, indent = 4)
```

```python
def agg(Nintendo=False):
    if Nintendo:
        gnin = df[df['Publisher'] == "Nintendo  "]
    else:
        gnin = df

    #Genre
    g1 = gnin.groupby(['Genre'])['Sales'].sum().reset_index().rename(columns={"Genre": "group", "Sales": "value"})
    #g1["opt"] = (g1['value'].rank(pct=True) + 1) / 2
    g1 = g1.to_dict('records')
    
    #Platform
    g2 = gnin[['NA_Sales',
            'PAL_Sales',
            'JP_Sales',
            'Other_Sales',
            'Total_Shipped']].rename(columns={'NA_Sales':'North America',
                                                'PAL_Sales':'Europe',
                                                'JP_Sales':'Japan',
                                                'Other_Sales':'Other',
                                                'Total_Shipped':'Unclassified'}).sum().reset_index().rename(columns={"index": "group", 0: "value"})
    #g2["opt"] = (g2['value'].rank(pct = True) + 1) / 2
    g2 = g2.to_dict('records')

    return g1, g2

g1, g2 = agg()
g3, g4 = agg(True)

for i, g in enumerate([g1, g2, g3, g4]):
    str = f"data{i + 1}{'_N' if i > 1 else ''}.json"
    with open(str, "w") as outfile:
        json.dump(g, outfile, indent = 4)
```
